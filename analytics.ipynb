```python
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
os.listdir()
```

```python
df = pd.read_csv('scrape-results_mt-07_16-02-2024.csv', index_col=[0])

df['build_year'] = df['build_year'].fillna(0)
df['build_year'] = df['build_year'].astype(int)

df['price'] = df['price'].str.extract(r'([\d]\.[\d][\d][\d])')
df['price'] = df['price'].str.replace('.','')
df['price'] = df['price'].fillna('0')
df['price'] = df['price'].astype(int)

df['mileage'] = df['mileage'].str.extract(r'([\d]+.[\d][\d][\d])')
df['mileage'] = df['mileage'].str.replace('.','')
df['mileage'] = df['mileage'].fillna('0')
df['mileage'] = df['mileage'].astype(int)


df['age'] = 2024-df['build_year']
```

```python
from scipy.stats import linregress
```

```python
def generate_line(xrange, linregress_results, npoints=10):

    xm, xM = xrange

    a = linregress_results.slope
    b = linregress_results.intercept
    
    xs = []
    ys = []
    xd = (xM - xm) / npoints

    x = xm
    
    for i in range(npoints):
        xs.append(x)
        y = a*x + b
        ys.append(y)
        x += xd

    return xs, ys
```

```python
build_years = [x for x in df['build_year'].unique() if x != 0]
build_years.sort()

fig, ax = plt.subplots(figsize=(8,6))

build_year_selection = [2024, 2022, 2020, 2018, 2016]

years = build_year_selection

for year in years:

    M = df['build_year'] == year
    
    xs = df['mileage'][m & M].values / 1000
    ys = df['price'][m & M].values / 1000
    
    sc = ax.scatter(xs, ys, s = 100, alpha=0.5, label='build year: {:}'.format(year))
    col = sc.get_facecolors()[0].tolist()

    try:
        res = linregress(xs,ys)
        lxs, lys = generate_line((xs.min(), xs.max()), res)
        ax.plot(lxs, lys, color = col)
    except:
        pass

ax.set_ylabel("Price (1000 EUR)")
ax.set_xlabel("Mileage (1000 km)")
ax.legend()
```
