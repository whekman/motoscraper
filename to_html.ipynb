```python
import pandas as pd
import os
```

```python
os.listdir()
```

```python
df = pd.read_csv('scrape-results_mt-07.csv', index_col=[0])
```

```python
df.head()
```

```python
df.columns
```

```python
cols = ['date', 'build_year', 'price', 'mileage','title']
```

```python
header = ""

for col in cols:
    header += "<th>{:}</th>".format(col)

header = "<tr>" + header + "</tr>"

lines = []
lines.append(header)
```

```python

for i in range(len(df)):

    line = ""

    for col in cols:
        
        line += "<td>{:}</td>".format(df.loc[i, col])
    
    line = "<tr>" + line + "</tr>"
    lines.append(line)
```

```python
lines
```

```python
with open("lines.txt","w") as f:
    f.write("\n".join(lines))
```
